```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import re
```

```python
page=requests.get("https://www.amazon.in/s?k=bags&crid=2M096C61O4MLT&qid=1653308124&sprefix=ba%2Caps%2C283&ref=sr_pg_1")
soup = BeautifulSoup(page.content,'html.parser')
page_title = soup.title.text
print(page_title)
```

```python
print("\nPrinting all the Product URL : \n")
Links = []
for Product_URL in soup.find_all('a', attrs={'href': re.compile("^https://")}):
    Links.append(Product_URL.get('href'))
    print(Product_URL.get('href'))  
```

```python
print("\nPrinting all the Product Names : \n")
for i in soup.find_all('span', {'class' : 'a-size-medium a-color-base a-text-normal'}):
    print(i.get_text()) 
```

```python
print("\nPrinting all the Product Price : \n")
for i in soup.find_all('span', {'class' : 'a-price-whole'}):
    print(i.get_text())     
```

```python
print("\nPrinting all the Product Rating : \n")
for i in soup.find_all('span', {'class' : 'a-icon-alt'}):
    print(i.get_text())     
```

```python
print("\nPrinting all the  Number of reviews : \n")
for i in soup.find_all('span', {'class' : 'a-size-base s-underline-text'}):
    print(i.get_text())     
```
